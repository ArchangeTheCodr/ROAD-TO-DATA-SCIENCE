```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
prets = pd.read_csv('https://raw.githubusercontent.com/OpenClassrooms-Student-Center/fr-4452741-decouvrez-les-librairies-python-pour-la-data-science/main/data/prets_final.csv')

prets.head()
```

```python
## 1. proportion de prêt par type de prêt.
sns.countplot(data=prets, x='type')
plt.rcParams.update({'font.weight' : 'bold', 'font.size' : 12})
plt.ylabel('Nombtre de prets')
plt.xlabel('Type de prets')
plt.title('Proportion de prêt par type de prêt')
sns.set_theme(style= 'darkgrid', palette='Set2')
plt.show()
```

```python
# bénéfice mensuel réalisé en fonction du revenu du client - prêts immobiliers
data = prets.loc[prets['type'] == 'immobilier']
sns.scatterplot(data=data, x='revenu', y='benefices', hue='ville')
plt.title('bénéfice mensuel réalisé en fonction du revenu du client')
plt.show()
```

```python
# La distribution des bénéfices réalisés
print('Avec un historigramme')
plt.hist(prets['benefices'])
plt.xlabel('Benefices en euros')
plt.ylabel('Nombres de clients')
plt.title('La distribution des bénéfices réalisés')
plt.show()
```

```python
display('Avec un diagramme a densite')
sns.kdeplot(prets['benefices'], fill=True, color="skyblue")
plt.title('KDE Plot')
plt.xlabel('Benefices')
plt.ylabel('Densité de probabilité')
plt.show()
```

```python
# bénéfice mensuel total réalisé par agence
sns.barplot(data=prets, x='ville', y='benefices', errorbar=None, estimator=sum, hue='type')
plt.title('bénéfice mensuel total réalisé par agence')
sns.set_theme()
```
